```python
def flat_func(arr):
    l = []
    for i in arr:
        if isinstance(i, list):
            l.extend(flat_func(i))
        else:
            l.append(i)
    return l

flat_func(arr)
```

```python
def endless_flatten(nd_list):
    if isinstance(nd_list, list):
        return [item for sublist in nd_list for item in endless_flatten(sublist)]
    else:
        return [nd_list]
    
endless_flatten(arr)
```

need to use .createOrReplaceTempView() because if you rerun the cell there will be a table already exists error when you use .createTempView()

```python
empdf = spark.read.load("employee.csv", format="csv", header = True,inferSchema = True)
empdf.printSchema()

officedf = spark.read.load("officecode.csv", format="csv", header = True  ,inferSchema = True)
empdf.printSchema()

empdf.createOrReplaceTempView("employee_table")
officedf.createOrReplaceTempView("office_table")

spark.sql("SELECT employee_table.employeeNumber, employee_table.firstName,employee_table.lastName,jobTitle, employee_table.officeCode, \
office_table.addressLine1, office_table.state  FROM employee_table \
join office_table ON office_table.officeCode = employee_table.officeCode").show(10)
```
