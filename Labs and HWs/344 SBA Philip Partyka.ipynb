```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['month_number'], df['total_profit'])
plt.title('Company profit per month')
plt.ylabel('Total profit')
plt.xlabel('Month number')
plt.grid(linestyle='--')
plt.ylim(0,500000)
```

```python
import numpy as np
```

```python
plt.plot(df['month_number'], df['total_profit'],
         color="red",
         linestyle="dashed",
         linewidth=3,
         marker='o',
         markerfacecolor="black",
         label="Profit data of last year")
plt.title('Company profit per month')
plt.ylabel('Total profit')
plt.xlabel('Month number')
plt.legend(loc="lower right")
plt.xticks(np.arange(min(df['month_number']), max(df['month_number'])+1, 1))
plt.yticks(np.arange(100000, 500001, 100000))

plt.show()
```

```python
df = df.transpose()
```

```python
df.columns = df.iloc[0]
```

```python
plt.plot(df.loc['facecream'], 
         label="Face Cream Sales Data",
         marker='o',
         linewidth=3)
plt.plot(df.loc['facewash'], 
         label="Face Wash Sales Data",
         marker='o',
         linewidth=3)
plt.plot(df.loc['toothpaste'], 
         label="Toothpaste Sales Data",
         marker='o',
         linewidth=3)
plt.plot(df.loc['bathingsoap'], 
         label="Bathing Soap Sales Data",
         marker='o',
         linewidth=3)
plt.plot(df.loc['shampoo'], 
         label="Shampoo Sales Data",
         marker='o',
         linewidth=3)
plt.plot(df.loc['moisturizer'], 
         label="Moisturizer Sales Data",
         marker='o',
         linewidth=3)
plt.legend(loc="upper left")
plt.xticks(df.columns)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales Data')
plt.ylabel('Sales units in number')
plt.xlabel('Month number')
plt.show()
```

```python
plt.figure(figsize=(6.5,4))
plt.scatter(df['month_number'], df['bathingsoap'], label='Bathing Soap Sales Data')
plt.grid(linestyle='--')
plt.xticks(np.arange(min(df['month_number']), max(df['month_number'])+1, 1))
plt.title('Bathing Soap Sales Data')
plt.ylabel('Number of units Sold')
plt.xlabel('Month Number')
plt.legend(loc="upper left")
plt.show()
```

```python
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]

plt.figure(figsize=(6,4))
plt.plot(weight, 
         height,
         color="green",
         linestyle="-.",
         linewidth=2,
         marker='o',
         markersize=10,
         label="Weight/Height data points")
plt.title('Average weight with respect to average height')
plt.ylabel('Height in cm')
plt.xlabel('Weight in kg')
plt.legend(loc="lower right")
plt.show()
```
